```python
import requests
from bs4 import BeautifulSoup

# URL of the main page
url = "https://results.eci.gov.in"

# Fetch the content of the page
response = requests.get(url)
html_content = response.content

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all links on the main page
links = soup.find_all('a')

# Print out all the links
for link in links:
    print(f"Link text: {link.text} - URL: {link.get('href')}")

# Example: Navigate to an inner page (adjust the href accordingly)
inner_page_url = url + "/inner_page_example.html"
response_inner = requests.get(inner_page_url)
inner_html_content = response_inner.content

# Parse the HTML content of the inner page
soup_inner = BeautifulSoup(inner_html_content, 'html.parser')

# Example: Extract specific data from the inner page
# You need to adjust this part based on the structure of the inner page
# For instance, extracting table data
```
